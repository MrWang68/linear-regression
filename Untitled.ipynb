```python
import numpy as np
import random
from matplotlib import pyplot as plt 
import matplotlib

def gen_line_data(sample_num=1000):
    """
    训练数据产生函数
    y = a*x1+b*x1^2+c*x1^3+d
    """
    x1 = np.random.rand(sample_num)*2-1
    #x2 = np.random.rand(sample_num)
    #x3 = np.random.rand(sample_num)
    x2 = np.power(x1,2)
    x3 = np.power(x1,3)
    x0 = np.ones(sample_num)
    abc = np.array([2, 4, 5, 6])
    x = np.concatenate(([x1], [x2], [x3], [x0]), axis=0).T #
    y = np.dot(x, abc.T)  # y 列向量#
    return x, y, abc

def gen_line_data_test(sample_num=100):
    """
    测试数据产生函数
    """
    x1 = np.random.rand(sample_num)*2-1
    #x2 = np.random.rand(sample_num)
    #x3 = np.random.rand(sample_num)
    x2 = np.power(x1,2)
    x3 = np.power(x1,3)
    x0 = np.ones(sample_num)
    x = np.concatenate(([x1], [x2], [x3], [x0]), axis=0).T #
    y = np.dot(x, np.array([2, 4 , 5, 6]).T)  # y 列向量
    return x, y

    
    
    
    
    

def mbgd(samples, y, abc,test_x ,test_y , step_size=0.1, max_iter_count=10000, batch_size=0.2):
    """
    MBGD（Mini-batch gradient descent）小批量梯度下降：每次迭代使用b组样本
    :param samples:训练集
    :param y:训练集标签
    :param abc:真实参数
    :param test_x:测试集
    :param test_y:测试集标签
    :param step_size:步长
    :param max_iter_count:迭代最高次数
    :param batch_size:取样比例
    """
    sample_num, dim = samples.shape#训练集数据行列数
    test_row, test_lis = test_x.shape#测试集数据行列数
    y = y.flatten() #训练集结果
    test_y = test_y.flatten() #测试集结果
    w = np.ones((dim,), dtype=np.float32)#参数
    sample_loss = 10#训练误差
    test_loss = 10#测试误差
    iter_count = 0#迭代次数
    loss_domain = np.zeros(max_iter_count, dtype=np.float32)#训练误差数组
    sample_loss_domain = np.zeros(max_iter_count, dtype=np.float32)#测试误差数组
    while test_loss > 0.01 and iter_count < max_iter_count:
        sample_loss = 0
        test_loss = 0
        error = np.zeros((dim,), dtype=np.float32)        

        index = random.sample(range(sample_num),
                              int(np.ceil(sample_num * batch_size)))
        batch_samples = samples[index]
        batch_y = y[index]

        for i in range(len(batch_samples)):#求偏导
            predict_y = np.dot(w.T, batch_samples[i])
            for j in range(dim):
                error[j] += (batch_y[i] - predict_y) * batch_samples[i][j]

        for j in range(dim):#更新参数
            w[j] += step_size * error[j] / sample_num

        for i in range(test_row):#损失函数计算（测试集）
            predict_y = np.dot(w.T, test_x[i])
            error = (1 / (test_row * 2)) * np.power((predict_y - test_y[i]), 2)
            test_loss += error
        loss_domain[iter_count] = test_loss
        for i in range(sample_num):#损失函数计算（训练集）
            predict_y = np.dot(w.T, samples[i])
            error = (1 / (sample_num * dim)) * np.power((predict_y - y[i]), 2)
            sample_loss += error
        sample_loss_domain[iter_count] = sample_loss
        
        #print("iter_count: ", iter_count, "the sample_loss:", sample_loss, "the test_loss:", test_loss)
        iter_count += 1
    x_domain = np.arange(iter_count)
    s_domain = sample_loss_domain[0:iter_count]
    y_domain = loss_domain[0:iter_count]
    plt.figure(1)
    plt.plot(x_domain,y_domain,label='test_loss')
    plt.plot(x_domain,s_domain,color='red',label='train_loss')
    plt.legend(loc = 'upper right')
    plt.figure(2)
    x1 = np.linspace(-1, 1, 100)
    x2 = np.power(x1,2)
    x3 = np.power(x1,3)
    x0 = np.ones(100)
    x = np.concatenate(([x1], [x2], [x3], [x0]), axis=0).T #
    y = np.dot(x, np.array([2, 4 , 5, 6]).T)  # y 列向量
    y_fit = np.dot(x, w.T)
    plt.plot(x1,y,label='real')
    plt.plot(x1,y_fit,color='red',label='fit')
    plt.legend(loc = 'upper right')
    return w

if __name__ == '__main__':
    samples, y ,abc= gen_line_data()#产生训练集
    test_x,test_y = gen_line_data_test()#产生测试集
    w = mbgd(samples, y, abc, test_x, test_y)#小批量梯度下降
    print(w)  
```
